```python
 import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
    return pd.read_csv(file_path)


file_path = "/adult.csv"


df = load_dataset(file_path)


df.head(5)
```

```python
# Categorical data - countplot
sns.countplot(data=df, x='sex')
plt.title('Count of individuals by Sex')
plt.show()

```

```python
# Pie chart for 'Sex'
df['sex'].value_counts().plot(kind='pie', autopct='%0.2f')
plt.title('Distribution of sex')
plt.show()
```

```python
# Numberical Value # Histogram treat as categorical data distribution of data
plt.hist(df['age'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['age'])
```

```python
# Boxplot # 5 number summary # median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier
sns.boxplot(df['age'])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load 'adult' dataset
df = pd.read_csv('/adult.csv')

# Box plot (Numerical - Categorical) # age by sex and survival status
sns.boxplot(x='sex', y='age', hue='income', data=df)
plt.title('Box Plot of Age by Sex and Income')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load 'adult' dataset
df = pd.read_csv('/adult.csv')

# Density plots for the distribution of 'age' based on 'income' status
sns.distplot(df[df['income'] == '<=50K']['age'], hist=False, label='Income <=50K')
sns.distplot(df[df['income'] == '>50K']['age'], hist=False, label='Income >50K')

plt.title('Density Plot of Age by Income Status')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load 'adult' dataset
df = pd.read_csv('/adult.csv')

# Heatmap (Categorical - Categorical) for 'income' vs 'age'
heatmap_data = pd.crosstab(df['income'], df['age'])
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='Blues')
plt.title('Heatmap for Income vs Age')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load 'test' dataset
df = pd.read_csv('/test.csv')

# Heatmap for 'passengers' vs 'month' vs 'year'
heatmap_data = df.pivot_table(values='Age', index='Type of Travel', columns='id')
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=.5)
plt.title('Heatmap for Age vs Type of Travel vs id')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load 'adult' dataset
df = pd.read_csv('/test.csv')

# Clustermap for selected columns
columns_to_cluster = ['Age', 'Gender', 'Customer Type']

# Pivot the table and create the clustermap
clustermap_data = df[columns_to_cluster].pivot_table(index='Gender', columns='Customer Type')
sns.clustermap(clustermap_data, cmap='coolwarm', annot=True, fmt='.1f', linewidths=.5)
plt.title('Clustermap for Selected Columns')
plt.show()



```
